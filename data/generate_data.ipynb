```python
import numpy as np
import pandas as pd
```

```python
# =========================
# 1) UBAH INI SESUKA KAMU
# =========================
N_DUMMY = 1000          # ubah jadi 100, 1000, dll
RANDOM_SEED = 42
CSV_OUT = "data_ash_dummy.csv"

# =========================
# 2) DATA ASLI (2 baris)
# =========================
BASE_ROWS = [
    # X1, X2, X3, y1, y2, y3, y4
    dict(Tinggi_Letusan_m=3000, Kec_Angin_km_jam=20, Arah_Angin_deg=90,
         Jarak_km=15, Luas_km2=45, Sudut_deg=30, Radius_km=5),
    dict(Tinggi_Letusan_m=5000, Kec_Angin_km_jam=40, Arah_Angin_deg=180,
         Jarak_km=40, Luas_km2=120, Sudut_deg=45, Radius_km=12),
]

# =========================
# 3) GENERATOR DUMMY
# =========================
def make_dummy_rows(n, seed=42):
    rng = np.random.default_rng(seed)
    rows = []

    for _ in range(n):
        tinggi = int(rng.integers(2000, 9001))  # meter
        kec = int(rng.integers(10, 91))         # km/jam
        arah = int(rng.integers(0, 360))        # derajat

        # Heuristik dummy (bukan fisika): hanya supaya pipeline jalan
        jarak = 0.0065 * tinggi + 0.55 * kec + rng.normal(0, 4)
        jarak = float(np.clip(jarak, 5, 140))

        luas = 0.03 * (jarak ** 2) + 0.02 * kec * jarak + rng.normal(0, 30)
        luas = float(np.clip(luas, 10, 900))

        sudut = 20 + 0.35 * kec + rng.normal(0, 5)
        sudut = float(np.clip(sudut, 15, 120))

        radius = 0.12 * jarak + rng.normal(0, 1.5)
        radius = float(np.clip(radius, 2, 40))

        rows.append(dict(
            Tinggi_Letusan_m=tinggi,
            Kec_Angin_km_jam=kec,
            Arah_Angin_deg=arah,
            Jarak_km=round(jarak, 2),
            Luas_km2=round(luas, 2),
            Sudut_deg=round(sudut, 2),
            Radius_km=round(radius, 2),
        ))

    return rows

# =========================
# 4) BUAT DATASET + SIMPAN
# =========================
all_rows = BASE_ROWS + make_dummy_rows(N_DUMMY, seed=RANDOM_SEED)
df = pd.DataFrame(all_rows)
df.to_csv(CSV_OUT, index=False)
```
