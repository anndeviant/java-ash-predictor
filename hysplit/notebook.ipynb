## 📝 Catatan Penting

### Format Tanggal HYSPLIT:
- **YY MM DD HH** → 2 digit tahun, bulan, hari, jam
- `25 03 01 00` = 1 Maret 2025, jam 00:00
- `01 12 25 12` = 25 Desember 2001, jam 12:00

### Format Nama File Meteorologi GDAS:
- **gdas1.mmmYY.wN**
  - `mmm` = bulan (jan, feb, mar, apr, dll)
  - `YY` = 2 digit tahun (25=2025, 24=2024)
  - `N` = minggu ke-N dalam bulan (1-5)

### Contoh:
- `gdas1.mar25.w1` → Maret 2025, minggu 1 (tanggal 1-7)
- `gdas1.mar25.w2` → Maret 2025, minggu 2 (tanggal 8-14)
- `gdas1.dec24.w4` → Desember 2024, minggu 4 (tanggal 22-28)

**Aturan penting:** Tanggal di file CONTROL harus SESUAI dengan periode data meteorologi yang digunakan!

## 📊 Analisis Data Spasial dari Hasil Simulasi

Dari file cdump, kita bisa ekstrak data:
- **Jarak_km**: Jarak terjauh abu dari sumber erupsi
- **Luas_km2**: Luas area yang terdampak abu
- **Sudut_deg**: Arah dominan penyebaran (0°=Utara, 90°=Timur)
- **Radius_km**: Radius maksimum penyebaran

```python
# ANALISIS SPASIAL HASIL DISPERSI ABU VULKANIK
import numpy as np
import os
from math import radians, cos, sin, asin, sqrt, atan2, degrees

def haversine(lat1, lon1, lat2, lon2):
    """
    Hitung jarak antara dua titik koordinat menggunakan formula Haversine
    Return: jarak dalam kilometer
    """
    R = 6371  # Radius bumi dalam km
    
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    
    return R * c

def calculate_bearing(lat1, lon1, lat2, lon2):
    """
    Hitung bearing/sudut dari titik 1 ke titik 2
    Return: sudut dalam derajat (0° = Utara, 90° = Timur)
    """
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlon = lon2 - lon1
    x = sin(dlon) * cos(lat2)
    y = cos(lat1) * sin(lat2) - sin(lat1) * cos(lat2) * cos(dlon)
    
    bearing = atan2(x, y)
    bearing = degrees(bearing)
    bearing = (bearing + 360) % 360
    
    return bearing

def analyze_cdump_simple():
    """
    Analisis sederhana file cdump untuk mendapatkan:
    - Jarak maksimum
    - Luas area terdampak
    - Arah penyebaran dominan
    - Radius penyebaran
    
    CATATAN: Ini analisis estimasi. Untuk hasil akurat, parse file cdump lengkap.
    """
    
    # Parameter sumber (dari simulasi sebelumnya)
    source_lat = -8.108  # Surabaya
    source_lon = 112.922
    
    # Cek apakah ada file cdump.txt (hasil konversi)
    cdump_txt = "F:/HYSPLIT/working/cdump.txt"
    
    if not os.path.exists(cdump_txt):
        print("⚠️  File cdump.txt tidak ditemukan!")
        print("    Jalankan cell konversi cdump terlebih dahulu.\n")
        print("📝  Atau saya akan berikan ESTIMASI berdasarkan parameter simulasi:")
        
        # Estimasi berdasarkan parameter umum HYSPLIT untuk dispersi abu
        height = 10000  # meter (10 km)
        duration = 24   # jam
        
        # Estimasi kasar (asumsi angin rata-rata 15 m/s)
        avg_wind_speed = 15  # m/s
        max_distance = (avg_wind_speed * duration * 3600) / 1000  # km
        
        # Estimasi luas (asumsi cone dengan sudut 30°)
        estimated_area = 3.14159 * (max_distance**2) * 0.25
        
        print("\n" + "="*60)
        print("ESTIMASI HASIL DISPERSI ABU (Tanpa Data Meteorologi Aktual)")
        print("="*60)
        print(f"Jarak_km:     {max_distance:.2f} km")
        print(f"Luas_km2:     {estimated_area:.2f} km²")
        print(f"Sudut_deg:    Tergantung arah angin dominan (perlu data met)")
        print(f"Radius_km:    {max_distance:.2f} km")
        print("="*60)
        print("\n⚠️  Ini hanya ESTIMASI! Untuk hasil akurat:")
        print("   1. Pastikan simulasi sudah dijalankan dengan data meteorologi")
        print("   2. Konversi cdump ke format text")
        print("   3. Jalankan analisis lengkap\n")
        
        return None
    
    # Jika file cdump.txt ada, parse dan analisis
    print("📖 Membaca file cdump.txt...\n")
    
    with open(cdump_txt, 'r') as f:
        lines = f.readlines()
    
    # Parse data (format cdump.txt bervariasi, ini contoh parsing sederhana)
    coordinates = []
    concentrations = []
    
    print("🔍 Menganalisis data konsentrasi abu...\n")
    
    # Threshold konsentrasi (untuk menentukan area terdampak)
    threshold = 0.01  # mg/m³ (sesuaikan dengan standar)
    
    # Parsing sederhana - sesuaikan dengan format aktual file Anda
    # Ini contoh, perlu disesuaikan dengan struktur file cdump.txt sebenarnya
    
    print("="*60)
    print("HASIL ANALISIS SPASIAL DISPERSI ABU")
    print("="*60)
    print(f"Sumber Erupsi: {source_lat}°, {source_lon}°")
    print(f"Lokasi: Surabaya, Jawa Timur")
    print(f"\nThreshold konsentrasi: {threshold} mg/m³")
    print("="*60)
    
    # Untuk analisis lengkap, perlu library khusus seperti PySPLIT
    print("\n💡 CATATAN:")
    print("   Untuk analisis detail, gunakan library PySPLIT")
    print("   atau parse manual file cdump dengan struktur lengkap")

# Jalankan analisis
analyze_cdump_simple()
```

```python
# ANALISIS LENGKAP - EKSTRAK DATA JARAK, LUAS, SUDUT, RADIUS
import numpy as np
import pandas as pd
import os
from math import radians, cos, sin, asin, sqrt, atan2, degrees

def parse_cdump_and_analyze(source_lat=-8.108, source_lon=112.922, threshold_conc=0.01):
    """
    Parse file cdump dan hitung metrik spasial
    
    Parameters:
    - source_lat, source_lon: Koordinat sumber erupsi
    - threshold_conc: Threshold konsentrasi minimum untuk dianggap terdampak (mg/m³)
    
    Returns:
    - DataFrame dengan metrik: Jarak_km, Luas_km2, Sudut_deg, Radius_km
    """
    
    cdump_txt = "F:/HYSPLIT/working/cdump.txt"
    
    if not os.path.exists(cdump_txt):
        print("❌ File cdump.txt tidak ada. Konversi file cdump dulu!\n")
        return None
    
    print("📊 ANALISIS DATA DISPERSI ABU VULKANIK")
    print("="*70)
    
    # Baca file
    with open(cdump_txt, 'r') as f:
        content = f.read()
    
    # Parsing data dari cdump.txt
    # Format umum cdump.txt HYSPLIT:
    # Baris berisi: Lat, Lon, Concentration, Time, Height, dll
    
    lines = content.strip().split('\n')
    
    data_points = []
    max_distance = 0
    max_lat, max_lon = source_lat, source_lon
    directions = []
    
    print(f"Sumber Erupsi: {source_lat:.4f}°N, {source_lon:.4f}°E")
    print(f"Threshold: {threshold_conc} mg/m³")
    print(f"\nMenganalisis {len(lines)} baris data...\n")
    
    # Coba parse data (struktur bisa berbeda tergantung versi HYSPLIT)
    affected_points = 0
    
    for line in lines[10:]:  # Skip header (biasanya 10 baris pertama)
        try:
            # Contoh parsing - sesuaikan dengan format sebenarnya
            parts = line.strip().split()
            if len(parts) >= 3:
                # Asumsi format: ... lat lon conc ...
                # Sesuaikan index sesuai file Anda
                try:
                    lat = float(parts[0])
                    lon = float(parts[1])
                    conc = float(parts[2]) if len(parts) > 2 else 0
                    
                    # Filter data yang valid
                    if -90 <= lat <= 90 and -180 <= lon <= 180 and conc > threshold_conc:
                        # Hitung jarak
                        distance = haversine(source_lat, source_lon, lat, lon)
                        
                        # Hitung bearing/arah
                        bearing = calculate_bearing(source_lat, source_lon, lat, lon)
                        
                        data_points.append({
                            'lat': lat,
                            'lon': lon,
                            'concentration': conc,
                            'distance_km': distance,
                            'bearing_deg': bearing
                        })
                        
                        affected_points += 1
                        
                        if distance > max_distance:
                            max_distance = distance
                            max_lat, max_lon = lat, lon
                        
                        directions.append(bearing)
                        
                except (ValueError, IndexError):
                    continue
        except:
            continue
    
    if len(data_points) == 0:
        print("⚠️  Tidak ada data konsentrasi yang melebihi threshold!")
        print("   Kemungkinan:")
        print("   1. Format file cdump.txt berbeda dari yang diharapkan")
        print("   2. Threshold terlalu tinggi")
        print("   3. Simulasi belum menghasilkan data dispersi\n")
        return None
    
    # Buat DataFrame
    df = pd.DataFrame(data_points)
    
    # HITUNG METRIK SPASIAL
    
    # 1. Jarak maksimum
    jarak_km = max_distance
    
    # 2. Radius (rata-rata jarak titik terdampak)
    radius_km = df['distance_km'].mean()
    
    # 3. Sudut dominan (arah penyebaran utama)
    if len(directions) > 0:
        # Hitung arah dominan dengan circular mean
        rad_directions = [radians(d) for d in directions]
        mean_sin = np.mean([sin(r) for r in rad_directions])
        mean_cos = np.mean([cos(r) for r in rad_directions])
        sudut_deg = degrees(atan2(mean_sin, mean_cos)) % 360
    else:
        sudut_deg = 0
    
    # 4. Luas area terdampak (estimasi dengan convex hull atau grid)
    # Metode sederhana: πr² dengan r = jarak maksimum
    # Untuk lebih akurat, gunakan convex hull
    luas_km2 = 3.14159 * (max_distance ** 2) * 0.5  # Faktor 0.5 karena tidak perfect circle
    
    # Alternatif: hitung luas dari jumlah grid cells
    # Asumsi resolusi grid 0.05° (dari CONTROL file) ≈ 5.5 km
    grid_resolution_km = 5.5
    grid_area = (grid_resolution_km ** 2)
    luas_km2_grid = affected_points * grid_area
    
    # Gunakan estimasi yang lebih konservatif
    luas_km2 = min(luas_km2, luas_km2_grid)
    
    # TAMPILKAN HASIL
    print("="*70)
    print("✅ HASIL ANALISIS SPASIAL")
    print("="*70)
    print(f"\n📏 Jarak_km:     {jarak_km:.2f} km")
    print(f"   (Titik terjauh: {max_lat:.4f}°, {max_lon:.4f}°)")
    print(f"\n📐 Sudut_deg:    {sudut_deg:.2f}°")
    
    # Konversi sudut ke arah mata angin
    if 337.5 <= sudut_deg or sudut_deg < 22.5:
        arah = "Utara"
    elif 22.5 <= sudut_deg < 67.5:
        arah = "Timur Laut"
    elif 67.5 <= sudut_deg < 112.5:
        arah = "Timur"
    elif 112.5 <= sudut_deg < 157.5:
        arah = "Tenggara"
    elif 157.5 <= sudut_deg < 202.5:
        arah = "Selatan"
    elif 202.5 <= sudut_deg < 247.5:
        arah = "Barat Daya"
    elif 247.5 <= sudut_deg < 292.5:
        arah = "Barat"
    else:
        arah = "Barat Laut"
    
    print(f"   (Arah dominan: {arah})")
    print(f"\n⭕ Radius_km:    {radius_km:.2f} km")
    print(f"   (Rata-rata jarak penyebaran)")
    print(f"\n🗺️  Luas_km2:     {luas_km2:.2f} km²")
    print(f"   (Estimasi area terdampak)")
    print(f"\n📍 Total titik terdampak: {affected_points}")
    print("="*70)
    
    # Buat summary DataFrame
    summary = pd.DataFrame({
        'Metrik': ['Jarak_km', 'Luas_km2', 'Sudut_deg', 'Radius_km'],
        'Nilai': [jarak_km, luas_km2, sudut_deg, radius_km],
        'Keterangan': [
            f'Jarak terjauh: {max_lat:.4f}°, {max_lon:.4f}°',
            f'Area terdampak abu > {threshold_conc} mg/m³',
            f'Arah dominan: {arah}',
            'Rata-rata jarak penyebaran'
        ]
    })
    
    print("\n📋 TABEL RINGKASAN:\n")
    print(summary.to_string(index=False))
    print("\n")
    
    return summary, df

# Fungsi helper (sudah didefinisikan di cell sebelumnya, duplikasi untuk standalone)
def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    return R * c

def calculate_bearing(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    x = sin(dlon) * cos(lat2)
    y = cos(lat1) * sin(lat2) - sin(lat1) * cos(lat2) * cos(dlon)
    bearing = atan2(x, y)
    bearing = degrees(bearing)
    bearing = (bearing + 360) % 360
    return bearing

# JALANKAN ANALISIS
result = parse_cdump_and_analyze(
    source_lat=-8.108,      # Surabaya
    source_lon=112.922,
    threshold_conc=0.01     # Threshold 0.01 mg/m³ (sesuaikan kebutuhan)
)

if result:
    summary_df, detail_df = result
    print("✅ Data berhasil diekstrak!")
    print("\nUntuk save ke file CSV:")
    print("   summary_df.to_csv('hasil_analisis_spasial.csv', index=False)")
```
