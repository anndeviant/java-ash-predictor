```python
import pandas as pd
from sklearn.impute import KNNImputer

# 1. Load data asli dari VONA
df = pd.read_csv('vona reports selain merapi_extracted.csv')

# 2. Pilih kolom fitur (Tinggi) dan kolom target yang kosong
cols_to_impute = ['tinggi_letusan_m', 'amplitudo', 'max_time']
data_subset = df[cols_to_impute].copy()

# 3. Inisialisasi KNNImputer (mencari 3 letusan paling mirip)
imputer = KNNImputer(n_neighbors=3)

# 4. Jalankan pengisian data
imputed_values = imputer.fit_transform(data_subset)

# 5. Masukkan kembali ke dataset asli
df[['tinggi_letusan_m', 'amplitudo', 'max_time']] = imputed_values

# Simpan hasil pembersihan ke file baru
df.to_csv('selainmerapi.csv', index=False)
```

```python
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

# Membaca dataset
# Asumsi nama file adalah merapi_dataset.csv berdasarkan path
file_path = 'java_hysplit_filled.csv'
df = pd.read_csv(file_path)

# Memilih kolom numerik yang relevan untuk KNN Imputer
# Kita menyertakan fitur penyebab (tinggi_letusan, angin, dll) untuk memprediksi fitur target yang kosong
numeric_cols = ['tinggi_letusan_m', 'kec_angin_km_jam', 'arah_angin_deg', 
                'amplitudo', 'duration', 'jarak_km', 'luas_km2', 'sudut_deg', 'radius_km']

# Standarisasi fitur sangat penting untuk KNN karena algoritma ini berbasis jarak (Euclidean)
scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df[numeric_cols]), columns=numeric_cols)

# Inisialisasi KNN Imputer
# Menggunakan n_neighbors=5 (nilai umum) dan weights='distance' agar titik yang lebih dekat 
# memberikan pengaruh yang lebih besar pada perhitungan rata-rata
imputer = KNNImputer(n_neighbors=5, weights='distance')
df_imputed_scaled = pd.DataFrame(imputer.fit_transform(df_scaled), columns=numeric_cols)

# Mengembalikan nilai yang telah diimputasi ke skala asli
df_imputed = pd.DataFrame(scaler.inverse_transform(df_imputed_scaled), columns=numeric_cols)

# Memasukkan kembali hasil imputasi ke dalam dataframe utama
df[numeric_cols] = df_imputed

# Menampilkan jumlah missing value setelah imputasi untuk verifikasi
print("Missing values setelah imputasi:\n", df[numeric_cols].isnull().sum())

# Menyimpan hasil dataset
output_file = 'java_hysplit_filleds.csv'
df.to_csv(output_file, index=False)
print(f"Data berhasi diimputasi dan disimpan ke {output_file}")
```
