# Downloader Data GDAS1 (2025 W1)

Notebook ini akan mengunduh file meteorologi minggu pertama (`w1`) untuk setiap bulan pada tahun 2025 secara otomatis dari arsip NOAA.

Data akan disimpan langsung ke direktori HYSPLIT komputer Anda, yaitu: `F:/HYSPLIT/metdata`.

```python
import os
import urllib.request

# Konfigurasi URL dan Direktori
BASE_URL = "https://www.ready.noaa.gov/data/archives/gdas1/"
DEST_DIR = "F:/HYSPLIT/metdata"
YEAR = "25"
MONTHS = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]

# Memastikan direktori tujuan tersedia
os.makedirs(DEST_DIR, exist_ok=True)

# Fungsi untuk memonitor progress (karena ukuran file cukup besar ~570MB)
def download_progress(count, block_size, total_size):
    if total_size > 0:
        percent = min(int(count * block_size * 100 / total_size), 100)
        print(f"\rProgress: {percent}% diunduh", end="")

print(f"Menyiapkan unduhan ke folder: {DEST_DIR}\n")

# Looping setiap bulan untuk mendownload w1 tahun 2025
for month in MONTHS:
    filename = f"gdas1.{month}{YEAR}.w1"
    url = BASE_URL + filename
    dest_path = os.path.join(DEST_DIR, filename)
    
    if os.path.exists(dest_path):
        print(f"[{filename}] Sudah tersedia di folder. Melewati...")
        continue
        
    print(f"\n[{filename}] Mengunduh dari {url} ...")
    try:
        urllib.request.urlretrieve(url, dest_path, reporthook=download_progress)
        print(f"\n[{filename}] Berhasil diunduh!")
    except Exception as e:
        print(f"\n[{filename}] Gagal mendownload: {e}")

print("\nSemua pemeriksaan dan pengunduhan selesai!")
```
